```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("BlackFriday.csv")

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind='bar',title='purchase distribution by age')
```

```python
df.groupby('Age').size()
```

```python
lst=[]

for i in df['Age'].unique():
    lst.append([i,df[df['Age']==i]['Product_ID'].nunique()])
    
data=pd.DataFrame(lst,columns=['Age','Products'])
```

```python
data.plot.bar(x='Age')
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind='bar',title='amount spend by age ')
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind = 'bar', title = 'Amount Spend by age')
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind = 'pie', autopct = '%0.1f')
```
